```python
import random
import numpy as np
import pandas as pd

import statsmodels.api as sm
from sklearn.metrics import roc_curve, auc
```

```python
# Create random data with missing values
df = pd.DataFrame(np.random.randint(0,100,size=(100000, 4)), columns=list('ABCD'))
df['E'] = df['D'].mask(np.random.random(len(df)) < .3)
df['target'] = np.where(df['A'] > 15, 1, 0)
df.head(5)
```

```python
# Create 500 series with imputed missing values
for i in range(500):
  df[f"U_{i}"] = df["E"].fillna(np.random.choice([1,2,3]))

# Create a list of a risk driver under review
var_col = [col for col in df if col.startswith('U_')]
```

```python
# Create statmodels logistic model
Xtrain = df[['A', 'B', 'C']]
ytrain = df[['target']]
  
# building the model and fitting the data
log_reg = sm.Logit(ytrain, Xtrain).fit()
```

```python
preds = []

for i in var_col:
  # performing predictions on the test dataset
  Xtest = df[['A', 'B', f'{i}']]
  yhat = log_reg.predict(Xtest)

  fpr, tpr, thresholds = roc_curve(df['target'], yhat)
  roc_auc = auc(fpr, tpr)
  preds.append(roc_auc)

preds
```
